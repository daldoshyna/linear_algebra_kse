#### Task 1
Write a function for each type of linear transformation:

- stretch(X,a,b),
- shear(X,a,b),
- reflection(X,a,b),
- rotation(X,t)

```python
def stretch(X, a, b):
    stretch_matrix = np.array([[a, 0], [0, b]])
    return X @ stretch_matrix

stretched_cat = stretch(kitty, 1.7, 0.7)

plt.figure(figsize=(7, 6), dpi=100)
plt.gca().set_aspect('equal', adjustable='box')
plt.plot(stretched_cat[:, 0], stretched_cat[:, 1], color='deeppink')
```

```python
stretched_cat
```

```python
def shear(X, a, b):
    shear_matrix = np.array([[1, a], [b, 1]])
    return X @ shear_matrix

sheared_cat = shear(kitty, 0.5, 0.0)

plt.figure(figsize=(7, 6), dpi=100)
plt.gca().set_aspect('equal', adjustable='box')
plt.plot(sheared_cat[:, 0], sheared_cat[:, 1], color='deeppink')
```

```python
sheared_cat
```

```python
def reflection(X, a, b):
    reflection_matrix = 1/(a**2 + b**2) * np.array([[(a**2 - b**2), 2*a*b], [2*a*b, (b**2 - a**2)]])
    return X @ reflection_matrix

reflected_cat = reflection(kitty, -1, 1)

plt.figure(figsize=(7, 6), dpi=100)
plt.gca().set_aspect('equal', adjustable='box')
plt.plot(reflected_cat[:, 0], reflected_cat[:, 1], color='deeppink')
```

```python
reflected_cat
```

```python
def rotation(X, theta):
    rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
    return X @ rotation_matrix

rotated_cat = rotation(kitty, np.pi/4)

plt.figure(figsize=(7, 6), dpi=100)
plt.gca().set_aspect('equal', adjustable='box')
plt.plot(rotated_cat[:, 0], rotated_cat[:, 1], color='deeppink')
```

```python
rotated_cat
```

#### Task 3

Write a function for each 3-D rotation
- rotate_xy (X, t),
- rotate_yz (X, t),
- rotate_xz (X, t)

Getting the 3d image:

```python
def read_off (filename):
    with open(filename, 'r') as f:
        if 'OFF' != f.readline().strip():
            raise ValueError('Not a valid OFF header')
        n_vents, n_faces, _ = map(int, f.readline().strip().split())
        verts = [list(map(float, f.readline().strip().split())) for _ in range(n_vents)]
        faces = [list(map(int, f.readline().strip().split()[1:])) for _ in range(n_faces)]
        return np.array(verts), faces

vertices, faces = read_off('/Users/dianaa/Documents/AI28 - 2/archive/ModelNet40/plant/test/plant_0247.off')

def plot_off(vertices, faces):
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot(111, projection='3d')
    mesh = Poly3DCollection([vertices[face] for face in faces],
                            alpha=0.3,
                            edgecolors='k')
    ax.add_collection3d(mesh)
    ax.scatter(vertices[:, 0], vertices[:, 1], vertices[:, 2], s=2, c='green')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    ax.auto_scale_xyz(vertices[:, 0], vertices[:, 1], vertices[:, 2])

plt.show()
plot_off(vertices, faces)
```

```python
def rotate_xy (vertices, theta):
    rotation_matrix = np.array([[np.cos(theta), -np.sin(theta), 0], [np.sin(theta), np.cos(theta), 0], [0, 0, 1]])
    return vertices @ rotation_matrix

rotated_xy = rotate_xy(vertices, np.pi)

plt.show()
plot_off(rotated_xy, faces)
```

```python
rotated_xy
```

```python
def rotate_yz (vertices, theta):
    rotation_matrix = np.array([[1, 0, 0], [0, np.cos(theta), -np.sin(theta)], [0, np.sin(theta), np.cos(theta)]])
    return vertices @ rotation_matrix

rotated_yz = rotate_yz(vertices, np.pi/6)

plt.show()
plot_off(rotated_yz, faces)
```

```python
rotated_yz
```

```python
def rotate_xz (vertices, theta):
    rotation_matrix = np.array([[np.cos(theta), 0, -np.sin(theta)], [0, 1, 0], [np.sin(theta), 0, np.cos(theta)]])
    return vertices @ rotation_matrix

rotated_xz = rotate_xz(vertices, np.pi/6)

plt.show()
plot_off(rotated_xz, faces)
```

```python
rotated_xz
```
